# Analisi rete con e senza rotazioni

```python
import matplotlib.pyplot as plt 
```

## Analisi senza rotazioni
### $R^{2}$ e $MAE$ in funzione della dimensione in $\AA$ della griglia

```python
physical_lengths = [7.75, 10, 12, 15, 20]
r2_phys_length = [0.9913947582244873, 0.9971, 0.9978269934654236, 0.9977307319641113, 0.997818112373352]
mae_phys_length = [0.02567794732749462, 0.0154, 0.013445980846881866, 0.013685869984328747, 0.013483372516930103]
```

```python
plt.rcParams['figure.dpi'] = 100
plt.rcParams['text.usetex'] = True

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8,9), sharex=True)
ax1.plot(physical_lengths, r2_phys_length, marker='^', ms=12, alpha=.8, label='Test R2')
ax1.legend(prop={'size': 14})
ax1.set_title("$R^{2}$ and $MAE$ test for different physical lengths of the grid \n trained for 80 epochs", fontsize=15)

ax2.plot(physical_lengths, mae_phys_length, marker='^', ms=12, alpha=.8, label='Test MAE', c='r')
ax2.legend(prop={'size': 14})
ax1.legend(prop={'size': 14})
ax2.set_xlabel("$L [\AA]$", fontsize=16)
ax1.set_ylabel("$R^2$", fontsize=16)
ax2.set_ylabel("$MAE~[Ha]$", fontsize=16)
for i in range(len(physical_lengths)):
    ax1.text(physical_lengths[i], r2_phys_length[i], f'{r2_phys_length[i]:f}', fontsize=14)
    ax2.text(physical_lengths[i], mae_phys_length[i], f'{mae_phys_length[i]:f}', fontsize=14)
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax2.tick_params(axis='x', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)
plt.show()

```

### $R^{2}$ e $MAE$ in funzione della $\gamma$ della gaussiana
#### (L della gaussiana fissata a 12 $\AA$)

```python
gamma = [0.2, 0.36, 0.375, 0.5]
r2_gamma = [0.9969136714935303, 0.9973235726356506, 0.9973328113555908, 0.9962165951728821]
mae_gamma = [0.01602778211236, 0.014650428667664528, 0.014888626523315907, 0.018]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8,9), sharex=True)
ax1.plot(gamma, r2_gamma, marker='^', ms=12, alpha=.8, label='Test R2')
ax1.legend(prop={'size': 14})
ax1.set_title("$R^{2}$ and $MAE$ test for different $\gamma$ of the gaussian potential \n trained for 80 epochs", fontsize=15)

ax2.plot(gamma, mae_gamma, marker='^', ms=12, alpha=.8, label='Test MAE', c='r')
ax2.legend(prop={'size': 14})
ax1.legend(prop={'size': 14})
ax2.set_xlabel("$\gamma$", fontsize=16)
ax1.set_ylabel("$R^2$", fontsize=16)
ax2.set_ylabel("$MAE~[Ha]$", fontsize=16)
for i in range(len(gamma)):
    ax1.text(gamma[i], r2_gamma[i], f'{r2_gamma[i]:f}', fontsize=14)
    ax2.text(gamma[i], mae_gamma[i], f'{mae_gamma[i]:f}', fontsize=14)
ax1.tick_params(axis='x', labelsize=12)
ax1.tick_params(axis='y', labelsize=12)
ax2.tick_params(axis='x', labelsize=12)
ax2.tick_params(axis='y', labelsize=12)
plt.show()

```

###  Plot $R^{2}$ e $MAE$ in funzione della dimensione del train test:
#### L della gaussiana fissata a 12 $\AA$; $\gamma$=0.375

```python
sizes = [10000, 20000, 40000, 70000, 110680]
r2_sizes = [0.9872957468032837, 0.9915481209754944, 0.9948416352272034, 0.9969044923782349, 0.9978269934654236]
mae_sizes = [0.03338715061545372, 0.026473574340343475, 0.02054522931575775, 0.015828654170036316, 0.013445980846881866]
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(8,9), sharex=True)
ax1.plot(sizes, r2_sizes, marker='^', ms=12, alpha=.8, label='Test R2')
ax1.legend(prop={'size': 14})
ax1.set_title("$R^{2}$ and $MAE$ test as a function of train sizes \n trained for 80 epochs", fontsize=15)

ax2.plot(sizes, mae_sizes, marker='^', ms=12, alpha=.8, label='Test MAE', c='r')
ax2.legend(prop={'size': 14})
ax1.legend(prop={'size': 14})
ax2.set_xlabel("$N$ instances", fontsize=16)
ax1.set_ylabel("$R^2$", fontsize=16)
ax2.set_ylabel("$MAE~[Ha]$", fontsize=16)
for i in range(len(sizes)):
    ax1.text(sizes[i], r2_sizes[i], f'{r2_sizes[i]:f}', fontsize=14)
    ax2.text(sizes[i], mae_sizes[i], f'{mae_sizes[i]:f}', fontsize=14)
ax1.set_xscale("log")
ax2.set_xscale("log")
ax1.set_yscale("log")
ax2.set_yscale("log")
ax1.tick_params(axis='x', labelsize=14)
ax1.tick_params(axis='y', labelsize=14)
ax2.tick_params(axis='x', labelsize=14)
ax2.tick_params(axis='y', labelsize=14)
plt.show()

```

## Analisi rotazioni
### (L della gaussiana fissata a 12 $\AA$;  $\gamma=0.375$)

```python
num_rotazioni = [0, 4, 8, 16, 80]
r2_rot = [0.9978269934654236, 0.9956477284431458, 0.9969268441200256, 0.9970179796218872, 0.9954217076301575]
mae_rot = [0.013445980846881866, 0.018952451646327972, 0.015967445448040962, 0.015551138669252396, 0.020004693418741226]
```
