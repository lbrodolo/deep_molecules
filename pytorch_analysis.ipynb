## Analisi risultati rete pytorch - potenziali generati on-the-fly

### Performance rispetto alla lunghezza fisica della griglia:


 In tutti i casi la griglia ha 32 punti per lato del cubo, lo spacing $s = p_{i} - p_{i-1}$ tra un punto e il successivo della griglia varia in base alla lunghezza fisica scelta.

```python
import matplotlib.pyplot as plt

physical_lengths = [7.75, 10, 12]
r2_test_l = [0.9931164383888245, 0.9971, 0.9973235726356506]
mae_test_l = [0.021873386576771736, 0.0154, 0.014650428667664528]
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 8), sharex=True)
plt.rcParams['figure.dpi'] = 100
plt.rcParams['text.usetex'] = True
ax[0].plot(physical_lengths, r2_test_l,marker='^', alpha=.8, label='R2 test', markersize=10, )
ax[1].plot(physical_lengths, mae_test_l, '-o', alpha=.8, label='MAE test', c='r')
for i in range(len(physical_lengths)):
    ax[0].text(physical_lengths[i], r2_test_l[i], f'{r2_test_l[i]:.4f}', size=12)
    ax[1].text(physical_lengths[i], mae_test_l[i], f'{mae_test_l[i]:.4f}', size=12)

ax[0].set_ylabel('$R2$', fontsize=16)
ax[1].set_ylabel('$MAE$ [$Ha$]', fontsize=16)
for i in range(2):
    ax[i].legend(prop={'size': 14})
    ax[i].tick_params(labelsize=12)
    # ax[i].set_xscale("log")
    # ax[i].set_yscale("log")
ax[0].set_title('R2 and MAE test for different physical lengths of the grid' + '\n' + 'trained for 80 epochs', size=16)
ax[1].set_xlabel('$L$ [$\AA$]', fontsize=16)
plt.savefig('r2_mae_test.png', dpi=300, facecolor='w',)
```

### Performance rispetto a $\gamma$ ($\sigma$ del potenziale gaussiano)

```python
gamma = [0.2, 0.36, 0.375, 0.5]
r2_test_g = [0.9969136714935303, 0.9973235726356506, 0.9973328113555908, 0.9962165951728821]
mae_test_g = [0.01602778211236, 0.014650428667664528, 0.014888626523315907, 0.018]
```

```python
fig, ax = plt.subplots(2, 1, figsize=(8, 8), sharex=True)

ax[0].plot(gamma, r2_test_g,marker='^', alpha=.8, label='R2 test', markersize=10, )
ax[1].plot(gamma, mae_test_g, '-o', alpha=.8, label='MAE test', c='r')
for i in range(len(gamma)):
    ax[0].text(gamma[i], r2_test_g[i], f'{r2_test_g[i]:.4f}', size=12)
    ax[1].text(gamma[i], mae_test_g[i], f'{mae_test_g[i]:.4f}', size=12)

ax[0].set_ylabel('$R2$', fontsize=16)
ax[1].set_ylabel('$MAE$ [$Ha$]', fontsize=16)
for i in range(2):
    ax[i].legend(prop={'size': 14})
    ax[i].tick_params(labelsize=12)
ax[0].set_title('R2 and MAE test for different $\gamma$ of the gaussian potential' + '\n' + 'trained for 80 epochs', size=16)
ax[1].set_xlabel('$\gamma$', fontsize=16)
plt.savefig('r2_mae_test.png', dpi=300, facecolor='w',)
```
